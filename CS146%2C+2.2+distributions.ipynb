Plot histograms and pdfs
For each of the gamma, normal, and beta distributions, use Python to do the following:
1. On the same set of axes, plot the distribution for 2 different values of its parameters.
2. Generate 100 samples from the distribution for one setting of its parameters and plot a
histogram of the sample values. On the same plot, compare the histogram to the probability
density function. The histogram will be noisy, but should align fairly well with the pdf and
look something like this:

```python
import matplotlib
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
gamma = stats.gamma(a=1, scale=1/0.5)
gamma1 = stats.gamma(a=2, scale=1/1)
# Plot the pdf
plt.figure(figsize=(8, 4))
x = np.linspace(0, 5)
plt.plot(x, gamma.pdf(x))
plt.plot(x, gamma1.pdf(x))
plt.title('PDF')
plt.show()
```

```python
# Plot a histogram of random samples
gsamples = gamma.rvs(size=10000)
plt.figure(figsize=(8, 4))
plt.hist(gsamples, range=[0, 5], density=True)
plt.plot(x, gamma.pdf(x), c="red")
plt.title('Histogram of Random Samples')

plt.show()
```

```python
normal = stats.norm(loc=2.5, scale=0.5)
normal1 = stats.norm(loc=2, scale=0.4)
# Plot the pdf
plt.figure(figsize=(8, 4))
x = np.linspace(0, 5)
plt.plot(x, normal.pdf(x))
plt.plot(x, normal1.pdf(x))
plt.title('PDF')
plt.show()
```

```python
nsamples = normal.rvs(size=1000000)
plt.figure(figsize=(8, 4))
plt.hist(nsamples, range=[1, 4], density=True)
plt.plot(x, normal.pdf(x), c="red")
plt.title('Histogram of Random Samples')

plt.show()
```

```python
beta = stats.beta(a=2, b=4)
beta1 = stats.beta(a=3, b=0.6)
# Plot the pdf
plt.figure(figsize=(8, 4))
x = np.linspace(0, 2)
plt.plot(x, beta.pdf(x))
plt.plot(x, beta1.pdf(x))
plt.title('PDF')
plt.show()
```

```python
bsamples = beta.rvs(size=1000000)
plt.figure(figsize=(8, 4))
plt.hist(bsamples, range=[0, 1.5], density=True)
plt.plot(x, beta.pdf(x), c="red")
plt.title('Histogram of Random Samples')

plt.show()
```
